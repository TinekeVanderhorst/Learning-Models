Model types

The first retrieval is based off the fan of the object alone
The second retrieval can be based off the container alone, the location alone, or both
Both is best so we use that

so far:
internal corrections rely on refraction, which depends on the maximum delay parameter
this is a simple popcorn model - one waiting time for everything

foil occurs only when a wrong container has been chosen
it also relies on the max delay parameter

This model uses the andersons foil model for internal correction, but only for correcting the retrievals of container/location.
